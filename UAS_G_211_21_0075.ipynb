```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
```

```python
# Dataset
dataset = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'],
    '1981-1990': [75, 99, 120, 255, 150, 170, 90, 45, 175, 365, 333, 298],
    '1991-2000': [98, 165, 149, 320, 330, 444, 473, 400, 251, 390, 455, 323],
    '2001-2010': [234, 422, 435, 138, 323, 144, 277, 30, 50, 75, 400, 266]
}

angka_bulan = [str(i + 1) for i in range(len(dataset['Month']))]

plt.figure(figsize=(16, 8))

x = np.arange(len(dataset['Month']))
width = 0.2
colors = ['black', 'grey', 'red']

for i, stasiun in enumerate(dataset.keys()):
    if stasiun != 'Month':
        plt.bar(x + (i * width), dataset[stasiun], width=width, label=stasiun, color=colors[i % len(colors)])

plt.xlabel('Month')
plt.ylabel('Rainfall (mm)')
plt.title('Statistical Data')

plt.xticks(x, angka_bulan)
plt.legend()
plt.show()
```

```python
# Dataset
data = {
    'jam (UTC)': ['00.00', '03.00', '06.00', '09.00', '12.00', '15.00', '18.00', '21.00'],
    'asimilasi': [30, 25, 24, 20, 21, 23, 27, 23],
    'observasi': [23, 25, 25, 30, 24, 28, 27, 24],
    'non Asimilasi': [23, 20, 20, 16, 18, 28, 28, 23]
}

df = pd.DataFrame(data)

df_melt = pd.melt(df, id_vars='jam (UTC)', var_name='Kategori', value_name='Jumlah')

plt.figure(figsize=(10, 4))
sns.lineplot(data=df_melt, x='jam (UTC)', y='Jumlah', hue='Kategori', palette={'asimilasi': 'grey', 'observasi': 'black', 'non Asimilasi': 'red'}, linewidth=3, markers=True, marker='o')

plt.xlabel('Jam (UTC)')
plt.ylabel('Celcius')
plt.title('Suhu Permukaan')

plt.legend()
plt.show()
```
